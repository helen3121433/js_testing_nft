```python

import pandas as pd

pandas_df = pd.read_csv('attributes_of_top_7_collections/cleaned_almost_famous_pandas.csv')
degenerate_df = pd.read_csv('attributes_of_top_7_collections/cleaned_degenerate_ape_academy.csv')
geckos_df = pd.read_csv('attributes_of_top_7_collections/cleaned_galactic_geckos.csv')
guildsaga_df = pd.read_csv('attributes_of_top_7_collections/cleaned_guild_saga_heroes.csv')
magicticket_df = pd.read_csv('attributes_of_top_7_collections/cleaned_magicticket.csv')
magnum_df = pd.read_csv('attributes_of_top_7_collections/cleaned_magnum_ai.csv')
monkelabs_df = pd.read_csv('attributes_of_top_7_collections/cleaned_monkelabs_dao.csv')
```

```python
pandas_df = pandas_df.sort_values(by=['price','img'], ascending = [False,True])
degenerate_df = degenerate_df.sort_values(by=['price','img'], ascending = [False,True])
geckos_df = geckos_df.sort_values(by=['price','img'], ascending = [False,True])
guildsaga_df = guildsaga_df.sort_values(by=['price','img'], ascending = [False,True])
magicticket_df = magicticket_df.sort_values(by=['price','img'], ascending = [False,True])
magnum_df = magnum_df.sort_values(by=['price','img'], ascending = [False,True])
monkelabs_df = monkelabs_df.sort_values(by=['price','img'], ascending = [False,True])
```

```python
dup_pandas_df = pandas_df.drop_duplicates(subset=['title','price'])
dup_degenerate_df = degenerate_df.drop_duplicates(subset=['title','price'])
dup_geckos_df = geckos_df.drop_duplicates(subset=['title','price'])
dup_guildsaga_df = guildsaga_df.drop_duplicates(subset=['title','price'])
dup_magicticket_df = magicticket_df.drop_duplicates(subset=['title','price'])
dup_magnum_df = magnum_df.drop_duplicates(subset=['title','price'])
dup_monkelabs_df = monkelabs_df.drop_duplicates(subset=['title','price'])
```

```python
new_pandas_df = dup_pandas_df.reset_index()
new_pandas_df = new_pandas_df.drop(labels='index', axis=1)
# new_pandas_df.head()

new_degenerate_df = dup_degenerate_df.reset_index()
new_degenerate_df = new_degenerate_df.drop(labels='index', axis=1)

new_geckos_df = dup_geckos_df.reset_index()
new_geckos_df = new_geckos_df.drop(labels='index', axis=1)

new_guildsaga_df = dup_guildsaga_df.reset_index()
new_guildsaga_df = new_guildsaga_df.drop(labels='index', axis=1)

new_magicticket_df = dup_magicticket_df.reset_index()
new_magicticket_df = new_magicticket_df.drop(labels='index', axis=1)

new_magnum_df = dup_magnum_df.reset_index()
new_magnum_df = new_magnum_df.drop(labels='index', axis=1)

new_monkelabs_df = dup_monkelabs_df.reset_index()
new_monkelabs_df = new_monkelabs_df.drop(labels='index', axis=1)
```
