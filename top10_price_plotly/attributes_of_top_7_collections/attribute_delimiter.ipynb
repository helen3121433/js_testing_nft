```python
import pandas as pd
import numpy as np
import re

all_df = pd.read_csv('(revised)nft_in_each_collections.csv', index_col=[0])
all_df.head()
```

```python
def get_trait_type (c_name):
    col_df = all_df[all_df['collectionName'] == c_name]
    col_df = col_df.reset_index()
    col_df['origin_attributes'] = col_df['attributes'].copy()
    draft_trait_list=[]
    for r in range(len(col_df['attributes'])):
                # find all name after "trait_type"
                regex_results = re.findall(r"'trait_type': '([\s\S]*?)'",col_df['attributes'][r])
                draft_trait_list.append(regex_results)
    #             print("regex_result: \n", regex_results, "\n")

    dup_trait_list=[]
    # remove the duplicate string in the list
    for i in draft_trait_list:
        if i not in dup_trait_list:
            dup_trait_list.append(i)
    print("dup_trait_list: \n", dup_trait_list, "\n")

    # found the longest string with trait_type names, that will be out final name list
    max_len = -1
    for ele in dup_trait_list:
        if len(ele) > max_len:
            max_len = len(ele)
            final_trait_list = ele

    print("final_trait_list: \n", final_trait_list,"\n","\n # of trait_type: ", len(final_trait_list))
    return final_trait_list , col_df
```

```python
def align_trait_type(final_trait_list, col_df):
    for cell_count in range(len(col_df['attributes'])):
        #check and see if current(i) trait_type existed
        for i, trait_name in enumerate(final_trait_list):
            check_str = f"'trait_type': '{trait_name}'"
            #if trait_type not in cell, add trait_type with value=none
            if check_str not in (col_df['attributes'][cell_count]):
                split_cell = col_df['attributes'][cell_count].split('}')
                # add the trait_type
                missing_trait_add = ", {'trait_type': '" + trait_name + "', 'value': 'none'"
                # insert the string as correct order as the final_trait_list
                split_cell.insert(i,missing_trait_add)
                # re-add the "}"
                trait_revised = '}'.join(split_cell)
                col_df['attributes'][cell_count] = trait_revised

    return col_df
```

```python
def pull_type_value(col_df):
    deli_att_list = col_df['attributes'].str.split(',',expand=False)
    # print(deli_att_list)
    # print(deli_att_list[1])
    # print("----split by comma(,) test deli_att_list[0]:---- \n",deli_att_list[0],"\n",len(deli_att_list),"\n")
    # print("----test it deli_att_list[0][2]: ---- \n", deli_att_list[0][2], "\n")

    #create a empty dataframe to store the revised attribute list
    new_att_list = pd.DataFrame(columns=['attributes'], index=range(len(col_df['attributes'])))
    type_string = "'trait_type':"
    value_string = "'value':"
    delimiter = ","

    # rejoin_string = []

    for row_count in range(len(deli_att_list)):
        rejoin_string = []
        type_value_list = []
        #print(deli_att_list[row_count])
        for position_count in range(len(deli_att_list[row_count])):        
            #print(deli_att_list[row_count][position_count])
            if(type_string) in (deli_att_list[row_count][position_count]):
                #print("type_string existed!")
                type_value_list.append(deli_att_list[row_count][position_count])   
            elif(value_string) in (deli_att_list[row_count][position_count]):
                type_value_list.append(deli_att_list[row_count][position_count])
                #print(type_value_list)
        rejoin_string = delimiter.join(type_value_list)
                #print(rejoin_string)
        new_att_list['attributes'].iloc[row_count] = rejoin_string
#         print(row_count, "-----\n", new_att_list['attributes'].iloc[row_count],"\n"
    
    display(new_att_list.head(10))
    return new_att_list
    
```

```python
def split_att(final_trait_list,new_att_list, col_df):
    att_df = new_att_list['attributes'].str.split(',', expand=True)
    print(final_trait_list, "\n")
    print("# of trait_type: " ,len(final_trait_list))
    print("att_df.shape: ", att_df.shape)
    display(att_df.head())
    split_att_df = att_df.loc[:, np.arange(len(att_df.columns)) %2 != 0]
    # split_att_df
            # use regular expression to clean ('value') and non word characters
    split_att_df = split_att_df.replace({'(\'value\')?(\W)':''}, regex=True)
    split_att_col_list = list(split_att_df)
    for col_count in range(len(split_att_col_list)):
        split_att_df.rename(columns={split_att_col_list[col_count]:final_trait_list[col_count]}, inplace=True)

    print("split_att_df: \n")
    display(split_att_df.head(10))
    col_df = col_df.merge(split_att_df, how='outer',left_index=True,right_index=True)
    print("col_df: \n")
    display(col_df.head(10))
    return col_df
```

## almost_famous_pandas

```python
almost_famous_pandas = 'almost_famous_pandas'
```

```python
pandas_list = get_trait_type(almost_famous_pandas)
pandas_trait = pandas_list[0]
pandas_df_1 = pandas_list[1]
print("------------------------------------")
print("\n", "pandas_trait: \n", pandas_trait, "\n")
display("pandas_df_1: \n", pandas_df_1.head(10))
```

```python
pandas_df_2 = align_trait_type(pandas_trait, pandas_df_1)
print("----------------------")
print("panda_df_2: \n")
display(pandas_df_2.head())
```

```python
print("------------------------")
print("pandas_att_list: \n")
pandas_att_list = pull_type_value(pandas_df_2)
```

```python
final_pandas_df = split_att(pandas_trait ,pandas_att_list, pandas_df_2)
```

```python
print("pandas_df_1: ",pandas_df_1.shape)
print("pandas_df_2: ",pandas_df_2.shape)
print("final_pandas_df: ",final_pandas_df.shape)
```

```python
final_pandas_df.dtypes
```

```python
final_pandas_df.to_csv('cleaned_almost_famous_pandas.csv')
```
